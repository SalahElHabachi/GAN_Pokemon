```python
import cv2
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import imageio

# Chargement des images
images = [cv2.imread(f'./generated_DCGAN/generated-images-{str(i).zfill(4)}.png') for i in range(1, 1000)]  # Charger toutes les images

# Slection des images  inclure dans l'animation
selected_images = [images[i] for i in range(0, 1000, 10)]  # Slectionner les images 1, 10, 20, etc.

# Cration de la figure et de l'axe pour l'animation
fig, ax = plt.subplots()
ax.set_axis_off()  # Dsactiver les axes pour obtenir une animation plus propre
im = ax.imshow(selected_images[0])

# Ajouter le numro de l'image
text = ax.text(10, 20, '', color='yellow', fontsize=12, fontweight='bold')

# Fonction d'animation
def update(frame):
    im.set_array(selected_images[frame])
    text.set_text(f'Image {frame * 10}')
    return im, text

# Crer l'animation
ani = FuncAnimation(fig, update, frames=len(selected_images), interval=100)

# Enregistrer l'animation en tant que GIF
gif_path = 'animation_DCGAN.gif'
ani.save(gif_path, writer='pillow')

plt.show()
```
